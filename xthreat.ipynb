```python
df = pd.read_csv('data1.csv')
df.head(20)
```

```python
df[["x","y"]] = df['location'].str.split(expand = True)
df[["endX","endY"]] = df['pass_end_location'].str.split(expand = True)
df.head(20)
```

```python

df['x'] = df.loc[:,'x'] = df.x.map(lambda x: x[1:-1]).astype(float)
df['y'] = df.loc[:,'y'] = df.y.map(lambda x: x[0:-1]).astype(float)

df['endX'] = df.loc[:,'endX'] = df.endX.map(lambda x: x[1:-1]).astype(float)
df['endY'] = df.loc[:,'endY'] = df.endY.map(lambda x: x[0:-1]).astype(float)

df = df.drop(['location', 'pass_end_location'], axis =1)
```

```python
xT = pd.read_csv('grid.csv', header  = None)
import pandas as pd
xT.to_csv('grid.csv')
```

```python
xT = np.array(xT)
xT
```

```python
xT_rows, xT_cols = xT.shape
```

```python
xT_cols
```

```python
df['x1_bin'] = pd.cut(df['x'], bins = xT_cols, labels = False)
df['y1_bin'] = pd.cut(df['y'], bins = xT_rows, labels = False)
df['x2_bin'] = pd.cut(df['endX'], bins = xT_cols, labels = False)
df['y2_bin'] = pd.cut(df['endY'], bins = xT_rows, labels = False)
```

```python
df.head(20)
```

```python
df['start_value'] = df[['x1_bin', 'y2_bin']].apply(lambda x: xT[x[1]] [x[0]], axis = 1)
df['end_value'] = df[['x2_bin', 'y2_bin']].apply(lambda x: xT[x[1]] [x[0]], axis = 1)
```

```python
df.head()
```

```python
df['xT'] = df['start_value'] - df['end_value']
```
